```python
# Dependencies
import requests
import json
from pprint import pprint
import pandas as pd
```

```python
# Collect data from FDIC API for each bank (From 2013 Q1 TO 2022 Q4)
chase_url="https://banks.data.fdic.gov/api/financials?filters=CERT%3A628&fields=CERT%2CREPDTE%2CASSET%2CROA%2CROE%2CEEFFR%2CNIMY%2CLNLSNTV%2CLNLSDEPR%2CDEPDASTR%2CRBC1AAJ%2CIDT1CER%2CIDT1RWAJR%2CRBCRWAJ&sort_by=REPDTE&sort_order=DESC&limit=40&offset=0&agg_term_fields=CERT&agg_limit=40&format=json&download=false&filename=data_file"
boa_url ="https://banks.data.fdic.gov/api/financials?filters=CERT%3A3510&fields=CERT%2CREPDTE%2CASSET%2CROA%2CROE%2CEEFFR%2CNIMY%2CLNLSNTV%2CLNLSDEPR%2CDEPDASTR%2CRBC1AAJ%2CIDT1CER%2CIDT1RWAJR%2CRBCRWAJ&sort_by=REPDTE&sort_order=DESC&limit=40&offset=0&agg_term_fields=CERT&agg_limit=40&format=json&download=false&filename=data_file"
wf_url="https://banks.data.fdic.gov/api/financials?filters=CERT%3A3511&fields=CERT%2CREPDTE%2CASSET%2CROA%2CROE%2CEEFFR%2CNIMY%2CLNLSNTV%2CLNLSDEPR%2CDEPDASTR%2CRBC1AAJ%2CIDT1CER%2CIDT1RWAJR%2CRBCRWAJ&sort_by=REPDTE&sort_order=DESC&limit=40&offset=0&agg_term_fields=CERT&agg_limit=40&format=json&download=false&filename=data_file"
gs_url="https://banks.data.fdic.gov/api/financials?filters=CERT%3A33124&fields=CERT%2CREPDTE%2CASSET%2CROA%2CROE%2CEEFFR%2CNIMY%2CLNLSNTV%2CLNLSDEPR%2CDEPDASTR%2CRBC1AAJ%2CIDT1CER%2CIDT1RWAJR%2CRBCRWAJ&sort_by=REPDTE&sort_order=DESC&limit=40&offset=0&agg_term_fields=CERT&agg_limit=40&format=json&download=false&filename=data_file"
citi_url="https://banks.data.fdic.gov/api/financials?filters=CERT%3A7213&fields=CERT%2CREPDTE%2CASSET%2CROA%2CROE%2CEEFFR%2CNIMY%2CLNLSNTV%2CLNLSDEPR%2CDEPDASTR%2CRBC1AAJ%2CIDT1CER%2CIDT1RWAJR%2CRBCRWAJ&sort_by=REPDTE&sort_order=DESC&limit=40&offset=0&agg_term_fields=CERT&agg_limit=40&format=json&download=false&filename=data_file"
fr_url="https://banks.data.fdic.gov/api/financials?filters=CERT%3A59017&fields=CERT%2CREPDTE%2CASSET%2CROA%2CROE%2CEEFFR%2CNIMY%2CLNLSNTV%2CLNLSDEPR%2CDEPDASTR%2CRBC1AAJ%2CIDT1CER%2CIDT1RWAJR%2CRBCRWAJ&sort_by=REPDTE&sort_order=DESC&limit=40&offset=0&agg_term_fields=CERT&agg_limit=40&format=json&download=false&filename=data_file"
```

```python
# Create search query, make request and store in json
response_chase = requests.get(chase_url).json()
response_boa = requests.get(boa_url).json()
response_wf = requests.get(wf_url).json()
response_gs = requests.get(gs_url).json()
response_citi = requests.get(citi_url).json()
response_fr = requests.get(fr_url).json()
```

```python
#test test test
first_republic_data = response_fr["data"]
first_republic_data
```

```python
# Create separate data frame of each bank for looping
chase_data = response_chase["data"]
boa_data = response_boa["data"]
wf_data = response_wf["data"]
gs_data = response_gs["data"]
citi_data = response_citi["data"]
fr_data = response_fr["data"]
```

```python
# Zip all data into one object
combined_data = zip(chase_data,boa_data,wf_data,gs_data,citi_data,fr_data)

# Set empty lists to hold all banks data
combined_roa = []
combined_roe = []
combined_nim = []
combined_lev = []
combined_cet1cr = []
combined_t1rbcr = []
combined_trbcr = []
combined_nlla = []
combined_nlld = []
combined_ddta = []
combined_eff = []

# Loop through each character
for chase,boa,wf,gs,citi,fr in combined_data:
    
    #Gather Return on Asset for each bank and store data into list
    combined_roa.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['ROA'], 'BoA': boa['data']['ROA'],
                    'Wells Fargo': wf['data']['ROA'], 'Goldman Sachs': gs['data']['ROA'],
                     'Citi': citi['data']['ROA'],'First Republic': fr['data']['ROA']
                    })
    
    #Gather Return on Equity for each bank and store data into list
    combined_roe.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['ROE'], 'BoA': boa['data']['ROE'],
                    'Wells Fargo': wf['data']['ROE'], 'Goldman Sachs': gs['data']['ROE'],
                     'Citi': citi['data']['ROE'],'First Republic': fr['data']['ROE']
                    })
    
    #Gather Net Interest Margin for each bank and store data into list
    combined_nim.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['NIMY'], 'BoA': boa['data']['NIMY'],
                    'Wells Fargo': wf['data']['NIMY'], 'Goldman Sachs': gs['data']['NIMY'],
                     'Citi': citi['data']['NIMY'],'First Republic': fr['data']['NIMY']
                    })
    
    #Gather Leverage(Core Capital Ratio) for each bank and store data into list
    combined_lev.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['RBC1AAJ'], 'BoA': boa['data']['RBC1AAJ'],
                    'Wells Fargo': wf['data']['RBC1AAJ'], 'Goldman Sachs': gs['data']['RBC1AAJ'],
                     'Citi': citi['data']['RBC1AAJ'],'First Republic': fr['data']['RBC1AAJ']
                    })
    
    #Gather Common Equity Tier 1 Capital Ratio for each bank and store data into list
    combined_cet1cr.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['IDT1CER'], 'BoA': boa['data']['IDT1CER'],
                    'Wells Fargo': wf['data']['IDT1CER'], 'Goldman Sachs': gs['data']['IDT1CER'],
                     'Citi': citi['data']['IDT1CER'],'First Republic': fr['data']['IDT1CER']
                    })
    
    #Gather Tier 1 Risk-Based Capital Ratio for each bank and store data into list
    combined_t1rbcr.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['IDT1RWAJR'], 'BoA': boa['data']['IDT1RWAJR'],
                    'Wells Fargo': wf['data']['IDT1RWAJR'], 'Goldman Sachs': gs['data']['IDT1RWAJR'],
                     'Citi': citi['data']['IDT1RWAJR'],'First Republic': fr['data']['IDT1RWAJR']
                    })
    
    #Gather Total Risk-Based Capital Ratio for each bank and store data into list
    combined_trbcr.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['RBCRWAJ'], 'BoA': boa['data']['RBCRWAJ'],
                    'Wells Fargo': wf['data']['RBCRWAJ'], 'Goldman Sachs': gs['data']['RBCRWAJ'],
                     'Citi': citi['data']['RBCRWAJ'],'First Republic': fr['data']['RBCRWAJ']
                    })
    
    #Gather Net Loans and Leases to Assets for each bank and store data into list
    combined_nlla.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['LNLSNTV'], 'BoA': boa['data']['LNLSNTV'],
                    'Wells Fargo': wf['data']['LNLSNTV'], 'Goldman Sachs': gs['data']['LNLSNTV'],
                     'Citi': citi['data']['LNLSNTV'],'First Republic': fr['data']['LNLSNTV']
                    })
    
    #Gather Net Loans and Leases to Deposit for each bank and store data into list
    combined_nlld.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['LNLSDEPR'], 'BoA': boa['data']['LNLSDEPR'],
                    'Wells Fargo': wf['data']['LNLSDEPR'], 'Goldman Sachs': gs['data']['LNLSDEPR'],
                     'Citi': citi['data']['LNLSDEPR'],'First Republic': fr['data']['LNLSDEPR']
                    })
    
    #Gather Domestic Deposits to Total Assets for each bank and store data into list
    combined_ddta.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['DEPDASTR'], 'BoA': boa['data']['DEPDASTR'],
                    'Wells Fargo': wf['data']['DEPDASTR'], 'Goldman Sachs': gs['data']['DEPDASTR'],
                     'Citi': citi['data']['DEPDASTR'],'First Republic': fr['data']['DEPDASTR']
                    })
    
    #Gather Efficiency Ratio for each bank and store data into list
    combined_eff.append({'Reporting Date': chase['data']['REPDTE'],
                     'Chase': chase['data']['EEFFR'], 'BoA': boa['data']['EEFFR'],
                    'Wells Fargo': wf['data']['EEFFR'], 'Goldman Sachs': gs['data']['EEFFR'],
                     'Citi': citi['data']['EEFFR'],'First Republic': fr['data']['EEFFR']
                    })
```

```python
# Create Return on Asset DataFrame               
roa_df = pd.DataFrame(combined_roa)

#Chage data type of reporting date and sort the values in ascending orders
ROA_df= roa_df.astype({"Reporting Date":"datetime64"}).sort_values(by="Reporting Date",ascending=True).reset_index(drop=True)

#Display Return on Asset DataFrame 
ROA_df.head()
```
